```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
for idx in range(140):
    df = pd.read_csv(f'/Users/shengfang/Desktop/bandgap bowing/direct_pixel/CsBiBr-CsSbI-1/Extracted_Band_Gaps_{idx}.csv',index_col = 0)
    grid = df.pivot(index='Y_coord', columns='X_coord', values='bandgap')

    cmap = mpl.colormaps.get_cmap('viridis')
    norm = mpl.colors.Normalize(vmin=1.7, vmax=2.7)
    fig, (ax_heatmap, ax_hist) = plt.subplots(1, 2, gridspec_kw={'width_ratios': [1, 1]}, figsize=(8, 3))

    # Plot the heatmap
    sns.heatmap(grid, cmap=cmap, norm=norm, ax=ax_heatmap)
    ax_heatmap.invert_yaxis()  # Invert y-axis of the heatmap
    ax_heatmap.set_xlabel('')  # Optionally remove the label
    ax_heatmap.set_ylabel('')

    # Remove the tick labels for heatmap
    ax_heatmap.set_xticklabels([])
    ax_heatmap.set_yticklabels([])
    ax_heatmap.tick_params(axis='both', which='both', length=0)

    # Plot the histogram
    # Create bins from min to max with a step of 0.05
    bins = np.arange(1.70, 2.70, 0.005)  # +0.05 to include the max value

    counts, edges = np.histogram(df['bandgap'].values, bins=bins, density=False)
    bar_heights = counts
#     bar_heights = counts / counts.sum()

    # Plot the histogram with the normalized heights
#     plt.bar(edges[:-1], bar_heights, width=np.diff(edges), edgecolor='black', align='edge')
    plt.bar(edges[:-1], bar_heights, width=np.diff(edges), align='edge')
    ax_hist.set_xlabel('Bandgap (eV)')
    ax_hist.set_ylabel('Probability')
    ax_hist.set_xlim(1.7,2.7)
#     ax_hist.set_ylim(0,1)
    ax_hist.tick_params(axis='both', direction='in', which = 'both', labelsize=12, width=1, length=6)
    ax_hist.minorticks_on()
    

    # Show the combined plot
    plt.subplots_adjust(wspace=0.5)  
    plt.savefig(f'/Users/shengfang/Desktop/bandgap bowing/direct_pixel/figures_single_droplet/CsBiBr-CsSbI-1/Extracted_Band_Gaps_{idx}.png', dpi=300, bbox_inches='tight')
    plt.show()
```
